```python
import pandas as pd
```

```python
sheet = pd.read_csv("eurovision_data.csv")
```

```python
sheet
```

```python
result = []

for (i, fr) in enumerate(sheet.columns):
    for (j, t) in enumerate(sheet.iloc[:, i]):
        # check if t is nan
        if t == t:
            result += [(fr.replace(' ', '_'), t.replace(' ', '_'))]

result
```

```python
to_write = pd.DataFrame.from_records(result)

to_write
```

```python
to_write.to_csv("eurovision_2021.csv", index=False, header=False)
```
